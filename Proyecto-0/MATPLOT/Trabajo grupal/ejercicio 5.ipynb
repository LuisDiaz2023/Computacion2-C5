### Ejercicio 5:

Crear una función que represente gráficamente esta expresión:

$$\sin(2 \pi f_1 t) + \sin(2 \pi f_2 t)$$

Siendo $f_1$ y $f_2$ argumentos de entrada (por defecto $10$ y $100$) y $t \in [0, 0.5]$. Además, debe mostrar:

* leyenda,
* título "Dos frecuencias",
* eje x "Tiempo ($t$)"

y usar algún estilo de los disponibles,

La grafica resultante es:

![ejercicio5](https://user-images.githubusercontent.com/19308295/117557892-748fb200-b03d-11eb-8009-7c2ca5718ccb.png)

```python
import numpy as np
import matplotlib.pyplot as plt

def OndaDobleFrecuencia(f1 = 10, f2 = 100):
    #Definir rango de tiempo t ∈[0, 0.5]
    t = np.linspace(0, 0.5, 500)
    
    #Definir los datos
    señal1 = np.sin(2 * np.pi * f1 * t)
    señal2 = np.sin(2 * np.pi * f2 * t)
    
    #Sumar las señales
    suma = señal1 + señal2
    
    #Graficar la señal sumada
    plt.plot(t, suma, label = f'f1={f1}, f2={f2}', color = "blue")
    
    #Características del gráfico
    plt.title("Dos frecuencias")
    plt.xlabel("Tiempo (t)")
    
    #Mostrar cuadrícula
    plt.grid(color = "white")
    
    #Mostrar leyenda
    plt.legend()
    
    #Características del fondo
    plt.gca().set_facecolor("lightgrey")
    
    #Mostrar gráfico
    plt.show()

#Resultado
OndaDobleFrecuencia()
```
