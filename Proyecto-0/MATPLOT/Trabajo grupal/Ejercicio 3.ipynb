Con parametros `s` y `c` podemos modificar el tamaño y el color respectivamente. Para el color, a cada valor numérico se le asigna un color a través de un **mapa de colores**; ese mapa se puede cambiar con el argumento `cmap`. Esa correspondencia se puede visualizar llamando a la función `colorbar`.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.randn(100)
sizes = 1000 * np.random.rand(100)

# Crear figura y ejes
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 8))

# Primer scatter plot
scatter1 = ax1.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='Blues')
fig.colorbar(scatter1, ax=ax1)

# Segundo scatter plot
scatter2 = ax2.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='Oranges')
fig.colorbar(scatter2, ax=ax2)

# Mostrar figura
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.randn(100)
sizes = 1000 * np.random.rand(100)

# Crear figura y ejes
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 8))

# Primer scatter plot con color verde
scatter1 = ax1.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='Greens')
fig.colorbar(scatter1, ax=ax1)


# Segundo scatter plot con color rosado
scatter2 = ax2.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='Reds')
fig.colorbar(scatter2, ax=ax2)


# Mostrar figura
plt.show()
```
