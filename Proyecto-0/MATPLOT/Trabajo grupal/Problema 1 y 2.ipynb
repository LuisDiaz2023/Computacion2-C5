### Ejercicio 1 

La tarea más habitual a la hora de trabajar con matplotlib es representar una función. Lo que tendremos que hacer es definir un dominio y evaluarla en dicho dominio. Por ejemplo:

## $$ f(x) = e^{-x^2} $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear el dominio
x = np.linspace(-1, 3, 400)

# Evaluar la función en el dominio
y = np.exp(-x ** 2)

# Graficar la función
plt.plot(x, y, label='Función $f(x)$')
plt.xlabel('Eje x')
plt.ylabel('$f(x)$')
plt.title('Gráfica de $f(x) = e^{-x^2}$')
plt.legend()
plt.grid(True)
plt.show()
```

### Ejercicio 2:

La función `plot` acepta una serie de argumentos para personalizar el aspecto de la función.

De la gráfica anterior debe salir asi:

![ejercicio2](https://user-images.githubusercontent.com/19308295/117557864-385c5180-b03d-11eb-8d48-088807aa5ea8.png)

Investigar :

$1 - f(x)$

https://matplotlib.org/2.0.2/api/pyplot_api.html

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la nueva función
def g(x):
    return 1 - np.exp(-x**2)

# Crear el dominio
x = np.linspace(-1, 3, 400)

# Evaluar la nueva función en el dominio
y = g(x)

# Graficar la nueva función
plt.plot(x, y, label='$1 - f(x)$')
plt.xlabel('Eje x')
plt.ylabel('$1 - f(x)$')
plt.title('Gráfica de $1 - f(x) = 1 - e^{-x^2}$')
plt.legend()
plt.grid(True)
plt.show()
```
