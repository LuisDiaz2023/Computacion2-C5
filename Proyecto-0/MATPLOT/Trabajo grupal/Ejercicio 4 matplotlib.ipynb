### Ejercicio 4:

Podemos crear figuras con varios sistemas de ejes, pasando a `subplot` el número de filas y de columnas. 

llegando a gráficas asi:

![ejercicio4](https://user-images.githubusercontent.com/19308295/117557885-6772c300-b03d-11eb-8781-48a7bbd37e92.png)

Realizamos el codigo para llegar a las graficas

```python
import matplotlib.pyplot as plt
import numpy as np

#Definimos los datos 
x = np.linspace(-1, 7, 100)
y1 = np.sin(x)
y2 = np.cos(x)

#Concretamos las medidas de las graficos (largo y ancho)
plt.figure(figsize=(10, 3))

#Grafica Funcion Seno
plt.subplot(2, 2, 1) # 2 filas, 2 columnas, primera gráfica
plt.plot(x, y1, 'black')
plt.title('Funcion Seno')

#Grafica Funcion Coseno
plt.subplot(2, 2, 2)  # 2 filas, 2 columnas, segunda gráfica
plt.plot(x, y2, 'black')
plt.title('Funcion Coseno')

#Ajustamos el diseño diseño para evitar superposiciones
plt.tight_layout()

#Imprimir las graficas
plt.show()
```
