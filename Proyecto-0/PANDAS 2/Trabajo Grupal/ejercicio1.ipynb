   10.-Cuál es la <strong>fecha máxima</strong> por país ?.

   12.-Cuantos registros hay por <strong>"continente"</strong>?
   
   13.-Cuantos <strong>nuevos casos hubo como máximo</strong> en una fecha reportada por cada país??

```python
import pandas as pd

# Cargar el DataFrame desde la URL
url = "https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-data.csv?raw=true"
df = pd.read_csv(url)

# 1. Qué países se encuentran en el DataFrame.
countries = df['location'].unique()
print("Países en el DataFrame:", countries)

# 2. Cuántos registros hay por continente.
records_by_continent = df.groupby('continent').size()
print("\nRegistros por continente:", records_by_continent)

# 3. Cuántos nuevos casos hubo como máximo en una fecha reportada por cada país.
max_new_cases_by_country = df.groupby('location')['new_cases'].max()
print("\nMáximos nuevos casos por país:", max_new_cases_by_country)

# 4. Cuál es el país con el mayor GDP per cápita (PBI).
country_max_gdp = df.loc[df['gdp_per_capita'].idxmax(), 'location']
print("\nPaís con el mayor GDP per cápita:", country_max_gdp)

# 5. Qué país tiene la mayor cantidad de fumadoras femeninas.
country_max_female_smokers = df.loc[df['female_smokers'].idxmax(), 'location']
print("\nPaís con la mayor cantidad de fumadoras femeninas:", country_max_female_smokers)

# 6. Qué país tiene la mayor cantidad de muertes por millón.
country_max_deaths_per_million = df.loc[df['total_deaths_per_million'].idxmax(), 'location']
print("\nPaís con la mayor cantidad de muertes por millón:", country_max_deaths_per_million)

# 7. Qué país ha tomado las medidas más estrictas hasta ahora.
country_strictest_measures = df.loc[df['stringency_index'].idxmax(), 'location']
print("\nPaís con las medidas más estrictas:", country_strictest_measures)

# 8. Cuál fue el país en alcanzar los 10.000 casos activos por primera vez.
df_sorted = df.dropna(subset=['total_cases']).sort_values(by='date')
country_first_10000_cases = df_sorted[df_sorted['total_cases'] >= 10000].iloc[0]['location']
print("\nPrimer país en alcanzar los 10.000 casos activos:", country_first_10000_cases)

# Crear un nuevo DataFrame solo con los países Perú y Guinea
df_filtered = df[df['location'].isin(['Peru', 'Guinea'])]

# 9. Cuántas filas hay por país.
rows_per_country = df_filtered.groupby('location').size()
print("\nFilas por país en el DataFrame filtrado:", rows_per_country)

# 10. Cuál es la fecha máxima por país.
max_date_per_country = df_filtered.groupby('location')['date'].max()
print("\nFecha máxima por país:", max_date_per_country)

# 11. Cuál es la fecha mínima por país.
min_date_per_country = df_filtered.groupby('location')['date'].min()
print("\nFecha mínima por país:", min_date_per_country)

# 12. Cuántos registros hay por "continente" en el nuevo DataFrame filtrado.
records_by_continent_filtered = df_filtered.groupby('continent').size()
print("\nRegistros por continente en el DataFrame filtrado:", records_by_continent_filtered)

# 13. Cuántos nuevos casos hubo como máximo en una fecha reportada por cada país en el DataFrame filtrado.
max_new_cases_by_country_filtered = df_filtered.groupby('location')['new_cases'].max()
print("\nMáximos nuevos casos por país en el DataFrame filtrado:", max_new_cases_by_country_filtered)
```
